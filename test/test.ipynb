```python
def histogram(img):
    
    pixels=[]

    for x in range(256):
        pixels.append(x)
  
    width,height=img.size
    counts=[]
  
    for i in pixels:
        temp=0
        for x in range(width):
            for y in range(height):
                if (img.getpixel((x,y))==i):
                    temp=temp+1

        counts.append(temp)
  

    return counts
```
