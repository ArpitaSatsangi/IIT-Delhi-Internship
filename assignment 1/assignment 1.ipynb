```python
from PIL import Image
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def histogram(img,n):
    
    pixels=[]

    for x in range(256):
        pixels.append(x)
  
  
    width,height=img.size
    counts=[]
  
  
    for i in pixels:
        temp=0
        for x in range(width):
            for y in range(height):
                if (img.getpixel((x,y))==i):
                    temp=temp+1

        counts.append(temp)
  
  
    
    plt.bar(pixels,counts)
    plt.xlabel("Intensity")
    plt.ylabel("Number of pixels")
    plt.title("Histogram "+n)
    plt.show()
```

```python
def equilization_histogram(imgGray, n):
    
    
    equ1 = cv2.equalizeHist(imgGray)
    dst1 = cv2.calcHist(equ1, [0], None, [256], [0,256])

    plt.hist(equ1.ravel(),256,[0,256])
    plt.xlabel("Intensity")
    plt.ylabel("Number of pixels")
    plt.title('Histogram '+n)
    plt.show()
  
```

# First image

```python
img1 = Image.open('10.jpg')
imgGray1 = img1.convert('L')
```

```python
histogram(imgGray1, '1.1')
```

```python
imgGray1  = cv2.imread("10.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
equilization_histogram(imgGray1, '1.2')
```

```python
equ1 = cv2.equalizeHist(imgGray1)
res1 = np.hstack((imgGray1, equ1))
 
cv2.imshow('1. BEFORE vs AFTER', res1)
cv2.imwrite('1_MORNING.jpg',res1)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
